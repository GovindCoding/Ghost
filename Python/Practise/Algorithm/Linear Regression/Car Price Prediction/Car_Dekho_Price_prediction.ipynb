```python
plt.figure(figsize=(20,30),facecolor='white')
plotnumber=1

for column in X:
    if plotnumber<=len(X.columns):
        ax=plt.subplot(5,2,plotnumber)
        plt.scatter(X[column],y)
        plt.xlabel(column)
        plt.ylabel('Selling price',fontsize=20)
    plotnumber+=1
plt.tight_layout()
```
