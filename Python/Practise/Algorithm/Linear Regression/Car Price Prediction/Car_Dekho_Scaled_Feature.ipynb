```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
df=pd.read_csv(r'C:\Users\Admin\Machine_Learning_Projects\Linear_Regression\Data\Car_Dekho\car data.csv')
```

```python
df.head()
```

```python
df=pd.get_dummies(df,columns=['Fuel_Type','Seller_Type','Transmission','Owner'],drop_first=True)
```

```python
df
```

```python
df['Age']=2020-df['Year']
df.head()
```

```python
df.drop(columns='Year',inplace=True)
```

```python
df.drop(columns='Car_Name',inplace=True)
```

```python
df.head(5)
```

```python
scaler=StandardScaler()
var=['Selling_Price','Present_Price','Kms_Driven','Age']
df[var]=scaler.fit_transform(df[var])
```

```python
df
```

```python
y=df['Selling_Price']
x=df.drop(columns='Selling_Price')
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state=355)
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
print('Training completed')
```

```python
reg.score(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
data = pd.DataFrame({"Actual": y_test, "Predicted":y_pred})
data
```

```python
reg.score(x_test,y_test)
```

```python
fig = plt.figure()
fig.suptitle('y_test vs predictions')

#X-label
plt.xlabel('y_test')

# Y-label
plt.ylabel('predcitions')
plt.scatter(y_test,y_pred)
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
R2 = metrics.r2_score(y_test,y_pred)
R2
```
