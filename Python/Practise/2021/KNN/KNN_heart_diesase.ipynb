```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/mvarshita/Machine-Learning-Projects/master/Heart%20Disease%20Predictor/heart.csv')
df
```

```python
df.describe().T
```

```python
df.isnull().sum()
```

```python
df['target'].value_counts()
```

```python
heart_patient=len(df[df['target']==1])
heart_no_patient=len(df[df['target']==0])
total_patient=len(df['target'])
Percentage_heart_patient=(heart_patient/total_patient)*100
Percentage_No_heart_patient=(heart_no_patient/total_patient)*100
print('Percentage of Individuals having heart disease:     {:.2f} % '.format(Percentage_heart_patient))
print('Percentage of Individuals Not having heart disease: {:.2f} % '.format(Percentage_No_heart_patient))
```

```python
plt.figure(figsize=(5,8))
sns.countplot(x=df['target'],palette="Blues")
plt.xticks(np.arange(2),('0=No,Risk','1=Yes,Risk'),rotation=0,fontsize=20)
plt.title("")
plt.xlabel("Heart Diease",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x=df['target'],hue=df['sex'])
plt.title("Heart diease frequency based on Gender",fontsize=20)
plt.xlabel("Heart Diease",fontsize=20)
plt.xticks(np.arange(2),('0=No,Risk','1=Yes,Risk'),rotation=0,fontsize=20)
plt.legend(["Female", "Male"],fontsize=20)
plt.show()
```

```python
males_heart_diesease=len(df[(df['sex']==1) & (df['target']==1)])
male_total=len(df[df['sex']==1])
female_heart_disease=len(df[(df['sex']==0) & (df['target']==1)])
female_total=len(df[df['sex']==0])
Per_Male=(males_heart_diesease/male_total)*100
Per_Female=(female_heart_disease/female_total)*100
print("Percentage of males with heart dieases:-  {:.2f}%".format(Per_Male))
print("Percentage of Females with heart dieases:- {:.2f}%".format(Per_Female))
```

```python
pd.crosstab(df['age'],df['target']).plot(kind='bar',figsize=(20,10))
plt.title('Heart Disease Frequency for Ages',fontsize=20)
plt.legend(["No Risk","Yes,Risk"],fontsize=20)
plt.xlabel('Age',fontsize=20)
plt.ylabel('Frequency')
```

```python
pd.crosstab(df['thalach'],df['target']).plot(kind='bar',figsize=(20,10))
plt.title('Heart Disease Frequency for maximum heart rate achieved',fontsize=20)
plt.legend(['No,risk',"Yes,risk"],fontsize=20)
plt.xlabel("\nThalch",fontsize=20)
```

```python
pd.crosstab(df['cp'],df['target']).plot(kind='bar',figsize=(15,10))
plt.legend(['No,risk',"Yes,risk"],fontsize=20)
plt.title("Heart diease frequency for various types of chest pain",fontsize=20)
plt.xticks(np.arange(4),("typical angina","atypical angina","non-anginal pain" , "asymptomatic"),rotation=0,fontsize=20)
plt.xlabel("\nType of Chest pain",fontsize=20)
```

```python
pd.crosstab(df['fbs'],df['target']).plot(kind='bar',figsize=(12,8))
plt.legend(['No,risk',"Yes,risk"],fontsize=20)
plt.title("Heart diease frequency for Fasting blood sugar",fontsize=20)
plt.xticks(np.arange(2),('fbs < 120 mg/dl', 'fbs > 120 mg/dl'), fontsize=20,rotation = 0)
plt.xlabel("\nFasting Blood Sugar",fontsize=20)
```

Data contains;

age - age in years
sex - (1 = male; 0 = female)
cp - chest pain type
trestbps - resting blood pressure (in mm Hg on admission to the hospital)
chol - serum cholestoral in mg/dl
fbs - (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
restecg - resting electrocardiographic results
thalach - maximum heart rate achieved
exang - exercise induced angina (1 = yes; 0 = no)
oldpeak - ST depression induced by exercise relative to rest
slope - the slope of the peak exercise ST segment
ca - number of major vessels (0-3) colored by flourosopy
thal - 3 = normal; 6 = fixed defect; 7 = reversable defect
target - have disease or not (1=yes, 0=no)

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.boxplot(data=df,width= 0.5,ax=ax,  fliersize=3)
```

```python
q=df['chol'].quantile(0.95)
df_cleaned=df[df['chol']<q]
```

```python
m=df_cleaned['trestbps'].quantile(0.95)
df_cleaned=df_cleaned[df_cleaned['trestbps']<m]
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.boxplot(data=df_cleaned,width= 0.5,ax=ax,  fliersize=3)
```

# Cholestrol Exploration

```python
sns.distplot(df['chol'])
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(25,8))
ax=sns.distplot(df_cleaned[df_cleaned['target']==0]['chol'],label="Do not have Heart Disease",ax=ax1)
ax.set(xlabel="Individual don't have heart disease")

ax=sns.distplot(df_cleaned[df_cleaned['target']==1]['chol'],label="Have Heart Disease",ax=ax2)
ax.set(xlabel="People have HEart Disease")
plt.show()
```

```python
# Get min, max and average of the Cholestoral of the people do not have heart diseas
print('Min cholestoral of people who do not have heart disease: ', min(df_cleaned[df_cleaned['target'] == 0]['chol']))
print('Max cholestoral of people who do not have heart disease: ', max(df_cleaned[df_cleaned['target'] == 0]['chol']))
print('Average cholestoral of people who do not have heart disease: ', df_cleaned[df_cleaned['target'] == 0]['chol'].mean())
```

```python
# Get min, max and average of the Cholestoral of the people who have heart diseas
print('Min cholestoral of people who  have heart disease: ', min(df_cleaned[df_cleaned['target'] == 1]['chol']))
print('Max cholestoral of people who  have heart disease: ', max(df_cleaned[df_cleaned['target'] == 1]['chol']))
print('Average cholestoral of people who have heart disease: ', df_cleaned[df_cleaned['target'] == 1]['chol'].mean())
```
