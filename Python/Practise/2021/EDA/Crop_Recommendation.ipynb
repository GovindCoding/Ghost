```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r'https://raw.githubusercontent.com/Gladiator07/Harvestify/master/Data-processed/crop_recommendation.csv')
df
```

```python
df.rename(columns={"label":"crop"},inplace=True)
```

```python
l=df['crop'].unique()
```

# Attributes Correlation 

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True,linewidths=0.4)
plt.title("Attributes Correlation \n",fontsize=20)
plt.show()
```

# Level Nitrogen content in soil suitable for specific crops

```python
plt.figure(figsize=(20,8))
sns.barplot(x='crop',y='N',data=df)
plt.title("Types of crop and suitable value of Nitrogen content in soil \n",fontsize=20)
plt.xticks(rotation=25)
plt.xlabel("Types of crops",fontsize=20)
plt.ylabel(" ratio of Nitrogen content in soil",fontsize=20)
plt.show()
```

# Level Phosphorus content in soil suitable for specific crops

```python
plt.figure(figsize=(20,8))
sns.barplot(x='crop',y='P',data=df)
plt.title("Types of crop and suitable value of Phosphorus content in soil \n",fontsize=20)
plt.xticks(rotation=25)
plt.xlabel("Types of crops",fontsize=20)
plt.ylabel(" ratio of Phosphorous content in soil",fontsize=20)
plt.show()
```

```python
sns.catplot(y='temperature',x='crop',data=df,aspect=4)
plt.title("Relation of various crops with Temperature range",fontsize=20)
plt.xticks(np.arange(22),(l),fontsize=15,rotation=20)
plt.xlabel("Type of crops ",fontsize=20)
plt.ylabel("Temperature",fontsize=20)
plt.show()
plt.show()
```
